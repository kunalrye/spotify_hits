```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/final_dataset.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

#Get the numeric columns

df = df.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', 'key', 'mode', 'year'])

#Encode important non-numeric features as ordinal numeric
le = LabelEncoder()
df['playlist_genre'] = le.fit_transform(df['playlist_genre'])
df['playlist_subgenre'] = le.fit_transform(df['playlist_subgenre'])

df = df.select_dtypes(include = 'number')
df.head()
```

```python
hits = df[df['Hit or Not'] == 1]
non_hits = df[df['Hit or Not'] == 0].sample(hits.shape[0], random_state = 42)
df = hits.append(non_hits)
```

```python
df.to_csv('data/final_subset.csv')
```
