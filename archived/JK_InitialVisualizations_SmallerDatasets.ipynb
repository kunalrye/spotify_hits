```python
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def make_heatmap(df, fig_size = None):
    """
    Given a dataframe df, this makes a heatmap of the dataframe
    It removes all non
    """
    df = df.select_dtypes(include = 'number')
    corr_mat = df.corr()
    
    if fig_size:
        plt.figure(figsize = fig_size)
    else:
        plt.figure()
    ax = plt.axes()
    sb.heatmap(corr_mat, ax = ax, annot = True)
    ax.set_title('Heatmap for Music Features')
    plt.show()
```

# Data by Artist

```python
df1 = pd.read_csv('data/data_by_artist.csv')
df1.head()
```

```python
df1.describe()
```

```python
#Min-max normalize popularity
df1['popularity'] = df1['popularity'] / 95.0
df1.describe()
```

```python
make_heatmap(df1, fig_size = (10,10))
```
