```python
tracks = pd.read_csv("data/data.csv")
genres = pd.read_csv('data/data_w_genres.csv')
top10 = pd.read_csv('data/top10s.csv', encoding = 'ISO-8859-1')
```

```python
genres = genres[['artists', 'genres']]
```

```python
features = tracks[['acousticness','danceability','duration_ms',
                   'energy', 'explicit', 'instrumentalness','key','liveness','loudness',
                  'mode', 'speechiness','tempo', 'valence']]
```

```python
# we should plot our features as a function of time
features.hist(xlabelsize = 5)
```

```python
from sklearn.cluster import KMeans
# Scaling
for col in ['loudness', 'tempo', 'key', 'duration_ms']:
    features[col] = ((features[col] - features[col].min()) / (features[col].max() - features[col].min()))

# Determining the cluster size
score_list = []
for i in range(2,15):
    kmeans_model = KMeans(n_clusters=i, random_state=3).fit(features)
    preds = kmeans_model.predict(features)
    score_list.append(kmeans_model.inertia_)

# Visualization of different cluster size performations
pd.DataFrame(score_list, index=range(2, 15)).plot(legend=False).set(xlabel="Cluster Size", ylabel="Inertia")
```

```python
# Training and Predicting
kmeans_model = KMeans(n_clusters=6, random_state=3).fit(features)
preds = kmeans_model.predict(features)
```
