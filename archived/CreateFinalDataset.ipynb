```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('data/all_features_kunal.csv')
df.head()
```

```python
df['year'] = 0

rows, cols = df.shape
for i in range(rows):
    date = df.loc[i,'track_album_release_date'].lower()
    year = int(re.findall(r"[0-9]{4}", date)[0])
    df.loc[i,'year'] = year

df.head()
```

```python
vals, counts = np.unique(df['Hit or Not'], return_counts = True)
dict(zip(vals, counts))
```

```python
df['artist_past_performance'] = 0

for j in range(rows):
    artist = df.loc[j,'track_artist']
    year = df.loc[j,'year']
    df_sub = df[df['track_artist'] == artist]
    df_sub = df_sub[df_sub['year'] < year]
    df_sub = df_sub[df_sub['Hit or Not'] == 1]
    df.loc[j,'artist_past_performance'] = df_sub.shape[0]

df.head(10)
```

```python
vals, counts = np.unique(df['artist_past_performance'], return_counts = True)
assert sum(counts) == rows
dict(zip(vals, counts))
```
