```python
import pandas as pd
```

```python
filepath = 'data/years/'

df = pd.read_json(filepath + '1950.json', 'records')
for year in range(1950, 2016):
    if year == 1950:
        df = pd.read_json(filepath + str(year) + '.json', 'records')
    else:
        temp = pd.read_json(filepath + str(year) + '.json','records')
        df = df.append(temp)
```

```python
df.to_csv(filepath + 'lyricsfromJSON.csv')
```
