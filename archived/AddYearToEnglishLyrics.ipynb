```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('data/tracks_w_lyrics_w_hits_en.csv')
df.head()
```

## Extract the Year from the Date

```python
df['year'] = 0

rows, cols = df.shape
for i in range(rows):
    date = df.iloc[i,9]
    year = int(re.findall(r"[0-9]{4}", date)[0])
    df.iloc[i,28] = year
```

## Write to a New CSV File 

```python
df.to_csv('data/tracks_w_lyrics_w_hits_year_en.csv')
```
